```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

matplotlib.style.use('ggplot')
#pd.set_option('display.mpl_style', 'default')
plt.rcParams['figure.figsize'] = (19, 5)
```

```python
yepss=pd.read_csv("311_data.csv")
```

```python
#  Plot counts of a specified column using Pandas
yepss.Borough.value_counts().plot(kind='barh')
```

```python
# Bar plot of median values for latitude 
yepss.groupby('Complaint Type')['Latitude'].agg(np.median).plot(kind = 'bar')
plt.ylim(40.5,40.9)
```

```python
# Bar plot of median values for longitude 
yepss.groupby('Complaint Type')['Longitude'].agg(np.median).plot(kind = 'bar')
plt.ylim(-73.8,-74.235)
```

```python
# Box plot example
# Had to install seaborn to ubuntu at this point and import it into this notebook as sns (see above)  
g = sns.boxplot(x="Borough", y="Latitude", hue="Agency Name", data=yepss, palette="Set1")
g.set(ylim=(40.5, 40.9))
```

```python
sns.palplot(sns.color_palette("Set1", 10))
```

```python
# Bar plot example
b = sns.barplot(x="Borough", y="Latitude", hue="Agency Name", data=yepss, palette="Set2");
b.set(ylim=(40.5, 40.9))
```

```python
# Another bar plot example
c = sns.barplot(x="Borough", y="Latitude", data=yepss, palette="Set1")
c.set(ylim=(40.5, 40.9))
```

```python
# Violin plot example
# despine() - Remove the top and right spines from plot(s)
sns.violinplot(yepss["Borough"], yepss["Latitude"], palette="BuPu_d").set_ylim(40.42, 40.95)
sns.despine(left=True)
```

```python
# Regression plot
sns.regplot("Latitude", "Longitude", data=yepss)
sns.despine()
```

```python
# Bar plot of median values
# Notice, as you may have already from precious examples on this notebook, that the first two arguments are for x and y axes repsectively
# Also, you can set x_lim and y_lim at the end of the *plot exam as is evident here and in another example above

sns.barplot("Complaint Type", "Longitude", data=yepss, palette="Set1", estimator=np.median).set_ylim(-73.8,-74.235)
plt.xticks(rotation=90)
```

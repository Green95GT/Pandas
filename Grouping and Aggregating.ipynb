```python
# Group data and obtain the mean
# [[0,2,3]] - display the 1st, 3rd, and 4th column only  
# Data is grouped by each unique Latitude and Longitude coordinate
# means are calculated for each column containing numerical data
# each unique coordinate corresponds to a unique key
# if two complaints from the same exact coordiate exists then the mean of the unique keys are calculated   
grouped1 = yeps.groupby(['Latitude','Longitude']).mean()
grouped1[[0,2,3]]
```

```python
# Group data and obtain the mean values
# mean values stem from each unique coordinate so you shall see only the Unique Key column containing mean values really
# The first column is the index and no means are calculated from here regardless of the data (unique or not) 
grpagg = yeps.groupby('X Coordinate (State Plane)').aggregate(np.mean)
grpagg[[0,2,13,14]]
```

```python
grpagg.shape
```

```python
# You could also do a similar calculation only with sums
# Rather than the unique keys being averaged they are summed
# [[]] is being used to choose which columns to display
grpsum = yeps.groupby('X Coordinate (State Plane)').aggregate(np.sum)
grpsum[[0,2,13,14]]
```

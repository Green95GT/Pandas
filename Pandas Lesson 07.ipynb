```python
States = ['NY', 'NY', 'NY', 'NY', 'FL', 'FL', 'GA', 'GA', 'FL', 'FL'] 
data = [1.0, 2, 3, 4, 5, 6, 7, 8, 9, 10]
```

```python
idx = pd.date_range('1/1/2012', periods=10, freq='MS')
# freq='MS' is month start frequency
# see --> https://pandas.pydata.org/pandas-docs/stable/timeseries.html#timeseries-offset-aliases
idx
```

```python
df1 = pd.DataFrame(data, index=idx, columns=['Revenue'])
df1
```

```python
df1['State'] = States
df1
```

```python
# Create a second dataframe
data2 = [10.0, 10.0, 9, 9, 8, 8, 7, 7, 6, 6]
idx2 = pd.date_range('1/1/2013', periods=10, freq='MS')
df2 = pd.DataFrame(data2, index=idx2, columns=['Revenue'])
df2['State'] = States
df2
```

```python
# Combine dataframes
df = pd.concat([df1,df2])
df
```

```python
# Ways of calculating outliers: Method 1

# make a copy of original df
newdf = df.copy()

# Create a new column 'x-mean' in newdf which is basically one step in a std dev calculation
# This is a list of the deviation from the mean calculated from the 'Revenue' column from df
newdf['(x - mean)'] = abs(newdf['Revenue'] - newdf['Revenue'].mean())

# Create a new column '1.96*std' which is a 95% cofidence interval from the 'Revenue' distribution
newdf['1.96*std'] = 1.96*newdf['Revenue'].std()

# Create a new column 'Outlier' which lists boolean values indecating if each data point falls outside the 95% confidence interval
newdf['Outlier'] = abs(newdf['Revenue'] - newdf['Revenue'].mean()) > 1.96*newdf['Revenue'].std()
newdf
```

```python
# Ways of calculating outliers: Method 2
# Group by item. In other words, in this method the mean and std dev is calculated for each state individually and outliers are
# calculated from the correspoding 95% confidence interval (for each state)

# make a copy of original df
newdf = df.copy()

State = newdf.groupby('State')

newdf['Outlier'] = State.transform( lambda x: abs(x-x.mean()) > 1.96*x.std() )
newdf['(x - mean)'] = State.transform( lambda x: abs(x-x.mean()) )
newdf['1.96*std'] = State.transform( lambda x: 1.96*x.std() )
newdf
```

```python
# Method 2a
# Group by multiple items, in the case by 'State' and the month in each state.
# For example: 2012-04-01 Revenue is 4. 2013-04-01 Revenue is 9. (4+9)/2 is 6.5.
# Hence the mean listed for those two years (same month and day) is 6.5

# make a copy of original df
newdf = df.copy()

StateMonth = newdf.groupby(['State', lambda x: x.month])

newdf['Outlier'] = StateMonth.transform( lambda x: abs(x-x.mean()) > 1.96*x.std() )
newdf['mean'] = StateMonth.transform( lambda x: x.mean() ) 
newdf['(x - mean)'] = StateMonth.transform( lambda x: abs(x-x.mean()) )
newdf['1.96*std'] = StateMonth.transform( lambda x: 1.96*x.std() )
newdf
```

```python
# Method 3. Basically method 2a in synthesized into a function.
# Group by multiple items

# make a copy of original df
newdf = df.copy()

StateMonth = newdf.groupby(['State', lambda x: x.month])

def s(group):
    group['x-Mean'] = abs(group['Revenue'] - group['Revenue'].mean())
    group['1.96*std'] = 1.96*group['Revenue'].std()  
    group['Outlier'] = abs(group['Revenue'] - group['Revenue'].mean()) > 1.96*group['Revenue'].std()
    return group

Newdf2 = StateMonth.apply(s)
Newdf2
```

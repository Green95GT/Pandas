```python
# New York Stock Exchange - Ford Motor Company (F)
# iex - investors exchange. Functions from pandas_datareader.data and pandas_datareader.wb extract data from various Internet sources into
# a pandas DataFrame.
# The Investors Exchange (IEX) provides a wide range of data through an API. Historical stock prices are available for up to 5 years.
# From the following url: https://pandas-datareader.readthedocs.io/en/latest/remote_data.html

f = web.DataReader('F', 'iex', datetime.datetime(2014, 1, 1), datetime.datetime(2017,1,1))
```

```python
f.loc['2016-03-11']
```

```python
# Plot the price and volume
# Had to upgrade Numexpr using the following command: 
# sudo -i
# pip3 install --user --upgrade Numexpr
# Also significant research was done to create this subplot as the original method suggusted in the tutorial did not work 

fig, axes = plt.subplots(nrows=2, ncols=1)
fig.tight_layout()
plt.subplots_adjust(hspace=0.35)
f['close'].plot(ax=axes[0]); axes[0].set_title('Ford Motor Company Price')
f['volume'].plot(ax=axes[1]); axes[1].set_title('Trading Volume')

plt.gcf().set_size_inches(15,8)
```

```python
# More plots
f.plot(subplots = True, figsize = (22, 11));
plt.legend(loc = 'best')
```

```python
# Not sure what a moving average is, however, I am just following the tutorial which called for this step
# Anyway *.rolling().mean() calculated the rolling mean of the closing price at every 20 days
mavg = f['close'].rolling(20).mean()
```

```python
# Moving average plot
f.close.plot(label='Yahoo')
mavg.plot(label='mavg')
plt.legend()
plt.gcf().set_size_inches(25,5)
```

```python
# KDE plot
# In statistics, kernel density estimation (KDE) is a non-parametric way to estimate the probability density function (PDF) of a random 
# variable. This function uses Gaussian kernels and includes automatic bandwith determination.
# See KDE plot for pandas also resaearch kernel density estimation in general

f.close.plot(kind='kde')
plt.gcf().set_size_inches(20,4)
```
